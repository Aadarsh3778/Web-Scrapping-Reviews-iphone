# Web Scrapping

### Problem Statement:-
- Scrap the reviews/comments from amazon site of iphone13 phone.

```python
# Importing Important Libraries
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
#amazon url
amazon_url = requests.get("https://www.amazon.in/Apple-iPhone-13-128GB-Midnight/product-reviews/B09G9HD6PD/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews").text
```

```python
soup = BeautifulSoup(amazon_url)
```

```python
coms = soup.find_all("div", class_ = "a-row a-spacing-none")
```

```python
'''
Creating a code which will scrap all the reviews , name and the title of revies given by the customers,
Creating 3 list variables to storte the data of name, commments and title,
Creating a for loop with range of pages by seeing in the site (i.e. pages were 25) to scrap thecomments and giving the link
creating the coms variable which contains the tags of a specific comments and also search for others using the same tag
by using find_all,
Again Creating a loop to extract all the comments in that specific page.

'''

full_name = []
title_review= []
full_review = []
for i in range(0,25):
    url = requests.get(f"https://www.amazon.in/Apple-iPhone-13-128GB-Midnight/product-reviews/B09G9HD6PD/ref=cm_cr_arp_d_paging_btm_2?ie=UTF8&pageNumber={i}&reviewerType=all_reviews").text
    soup = BeautifulSoup(url)   
    coms = soup.find_all("div", class_ = "a-row a-spacing-none")
    
    for com in coms:
        name = com.find('span', class_ = "a-profile-name").text    
        rev = com.find("a", class_ ="a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold").text.replace("\n","")
        full_rev = com.find('span', class_ = "a-size-base review-text review-text-content").text.replace("\n","")

        full_name.append(name)
        title_review.append(rev)
        full_review.append(full_rev)


        print(f"Name: {name}")
        print(f"Title: {rev}")    
        print(f"Full Review: {full_rev}")
        print("="*100)
```

```python
len(full_name), len(title_review), len(full_review)
```

```python
df = pd.DataFrame({"Name" : full_name, "Comment_Title" : title_review, "Comments" : full_review})
```
